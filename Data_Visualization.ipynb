### Groupby

```python
covid.groupby("Country/Region")
```

Some aggregate function of `groupby`:
- `mean()`: Compute mean of groups
- `sum()`: Compute sum of group values
- `size()`: Compute group sizes
- `count()`: Compute count of group
- `std()`: Standard deviation of groups
- `var()`: Compute variance of groups
- `sem()`: Standard error of the mean of groups
- `describe()`: Generates descriptive statistics
- `first()`: Compute first of group values
- `last()`: Compute last of group values
- `nth()` : Take nth value, or a subset if n is a list
- `min()`: Compute min of group values
- `max()`: Compute max of group values

```python
covid.groupby("Country/Region").mean()[["Confirmed"]]
```

```python
covid.groupby("Country/Region").mean().sort_values("Confirmed", ascending=False)
```

```python
covid.groupby("Country/Region")[["Recovered"]].mean()
```

```python
covid.groupby("Country/Region")[["Recovered"]].mean().\
    sort_values(by="Recovered", ascending=False)
```

### Group data using index in a pivot table
* pivot_table requires a data and an index parameter
* data is the Pandas dataframe you pass to the function
* index is the feature that allows you to group your data. The index feature will appear as an index in the resultant table

```python
table1 = pd.pivot_table(data=covid, index=['WHO Region'])
table1
```

```python
table1 = table1[["Active", "Confirmed", "Deaths", "Recovered"]]
table1
```

```python
table1.plot(kind = 'bar')
plt.show()
```

```python
table1.plot.bar()
plt.show()
```

```python
table2 = table1[["Active", "Deaths", "Recovered"]]
table2
```

```python
table2.plot.bar(stacked=True)
plt.show()
```

```python
table3 = pd.pivot_table(covid, index=['WHO Region','DeathsOccur?'])
table3
```

จำนวนผู้ป่วยตามภูมิภาคต่าง ๆ นับแต่ละวัน

```python
count_cases_by_region = pd.pivot_table(covid,index=['Date'],
                       columns=["WHO Region"],
                       values=["Confirmed"],
                       aggfunc=np.sum)
count_cases_by_region
```

```python
count_cases_by_region.plot.line()
```

```python
regions = count_cases_by_region.columns
regions
```

```python
regions = [i[1] for i in list(regions)]
regions
```

```python
count_cases_by_region.plot.line(xlabel="",
                                ylabel="confirmed",
                                title="Confirmed by Date")
plt.legend(regions)
```
